```python
class Library:
    def __init__(self):
        self.books = {
            "Python": 5,
            "Java": 3,
            "C++": 2,
            "Ruby": 1
        }
        self.users = {
            "user1": "password1",
            "user2": "password2"
        }
        self.rented_books = {}
        self.username = ""
        self.password = ""
    def login(self, username, password):
        if username in self.users and self.users[username] == password:
            self.username = username
            self.password = password
            return True
        else:
            return False

    def logout(self):
        self.username = ""
        self.password = ""

    def check_books(self):
        for book in self.books:
            print(f"{book} - {self.books[book]}")

    def rent_book(self, book):
        if book in self.books and self.books[book] > 0:
            self.books[book] -= 1
            self.rented_books[self.username] = book
            print(f"Book {book} rented by {self.username}")
        else:
            print(f"Book {book} not available")

    def payment(self):
        if self.username in self.rented_books:
            book = self.rented_books[self.username]
            self.books[book] += 1
            del self.rented_books[self.username]
            print(f"Book {book} returned by {self.username}")
        else:
            print(f"No book rented by {self.username}")
```

```python
# Sample usage of the class
library = Library()

while True:
    print("================================")
    print("Menu")
    print("1.Login")
    print("2.Browse Books")
    print("3.Rent Book")
    print("4.Logout")
    print("==================================")
    choice=int(input("Enter any choice:"))
    if(choice==1):
        uname=input("Enter username:")
        pswd = input("Enter Password:")
        if library.login(uname,pswd):
            print("Login Successful")
        else:
            print("Invalid login credentials")
    elif choice==2:
        print("Getting you the list of books available")
        library.check_books()
    elif choice==3:
        r=input("Enter book name to rent")
        library.rent_book(r)
    elif choice==4:
        print("Log out success")
        break
```
